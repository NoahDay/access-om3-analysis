````python
import re
import glob
import pandas as pd
from pathlib import Path

def hms_to_hours(hms):
    """Convert HH:MM:SS to hours (float)."""
    h, m, s = map(float, hms.split(':'))
    return h + m/60 + s/3600

def parse_memory(mem_str):
    """Convert memory strings like '122.64GB' or '8.16MB' to GB."""
    num, unit = re.match(r"([\d.]+)([A-Za-z]+)", mem_str).groups()
    num = float(num)
    factor = {'GB': 1, 'MB': 1/1024, 'KB': 1/(1024**2)}.get(unit, 1)
    return num * factor

def parse_pbs_block(block):
    """Extract metrics from a single PBS resource usage block."""
    try:
        return {
            "JobID": re.search(r"Job Id:\s+([\w.-]+)", block).group(1),
            "Date": re.search(r"Resource Usage on ([\d-]+ \d+:\d+:\d+)", block).group(1),
            "ExitStatus": int(re.search(r"Exit Status:\s+(\d+)", block).group(1)),
            "ServiceUnits": float(re.search(r"Service Units:\s+([\d.]+)", block).group(1)),
            "NCPUs": int(re.search(r"NCPUs Requested:\s+(\d+)", block).group(1)),
            "CPUTime_hr": hms_to_hours(re.search(r"CPU Time Used:\s+(\d+:\d+:\d+)", block).group(1)),
            "MemUsed_GB": parse_memory(re.search(r"Memory Used:\s+([\d.]+[A-Za-z]+)", block).group(1)),
            "Walltime_hr": hms_to_hours(re.search(r"Walltime Used:\s+(\d+:\d+:\d+)", block).group(1)),
            "JobFS_GB": parse_memory(re.search(r"JobFS used:\s+([\d.]+[A-Za-z]+)", block).group(1)),
        }
    except AttributeError:
        return None


# --- Input models and their directories ---
models = ["48_proc", "96_proc", "192_proc", "240_proc", "288_proc", "480_proc_96_cice"]
base_dir = "/g/data/ps29/nd0349/access-om3/configurations/mcw/scaling_performance"

all_records = []

# Loop over all model directories
for model in models:
    model_dir = f"{base_dir}/{model}"
    if model == "480_proc_96_cice":
        model = "480_proc"
    files = glob.glob(f"{model_dir}/{model}.o*")

    for file in files:
        file_path = Path(file)
        
        try:
            with open(file) as f:
                text = f.read()
        except FileNotFoundError:
            print(f"⚠️ File not found: {file}")
            continue

        # Find PBS resource usage blocks
        blocks = re.findall(r"=+\n\s+Resource Usage on[\s\S]+?=+\n", text)
        records = [parse_pbs_block(block) for block in blocks]
        records = [r for r in records if r is not None]

        # Add model label to each record
        for r in records:
            r["Model"] = model

        all_records.extend(records)

# --- Combine into DataFrame ---
if all_records:
    df = pd.DataFrame(all_records)
    cols = ["Model"] + [c for c in df.columns if c != "Model"]
    df = df[cols]
else:
    df = pd.DataFrame(columns=[
        "Model", "JobID", "Date", "ExitStatus", "ServiceUnits", "NCPUs",
        "CPUTime_hr", "MemUsed_GB", "Walltime_hr", "JobFS_GB"
    ])

df = df[df["ServiceUnits"] >= 10]
df
````

````python
def intFloeSize(ds, var):
    tmp_var = var[0:-3]
    puny = 10**-12
    int_var = ds[tmp_var].isel(nf=1)
    int_var = int_var.where(ds['aice'] > puny, drop=False)
    NFSD = ds.NFSD.values
    floe_binwidth, _, _ = ciceBinWidths(NFSD)
    ds_out = ds.copy()
    # Put in temporary data
    ds_out[var] = ds_out[tmp_var].isel(nf=0)
    ds_out[var][:] = 0.0
    int_values = np.zeros(int_var.shape)
    tmp_data = np.zeros(ds['aice'].shape)
    ds_tmp = ds.where(ds['aice'] > puny, drop=False)
    
    tmp_data = (ds_tmp[tmp_var]*ds_tmp.NFSD/ds_tmp.aice).sum(dim='nf')
    ds_out[var].values = tmp_data
    
    # Set attributes
    parts = ds[tmp_var].attrs['long_name'].split(':')
    part1 = parts[0]
    part2 = parts[1]
    ds_out[var].attrs['long_name'] = part1[:-3] + 'rep. radius:' + part2
    ds_out['dafsd_wave_ra'].attrs['units'] = 'm/timestep'
    return ds_out

track_number = 201626905
df_cyclone = pd.read_csv(netcdfpath+'extratropical_v0.2.csv') 
df_single_track = df_cyclone.copy()
index = df_single_track[(df_single_track['# track'] != track_number)].index
df_single_track.drop(index , inplace=True)
cice_output_file = netcdfpath+"{}_CICE_output.nc".format(track_number)
ds = xr.open_dataset(cice_output_file).isel(time=slice(0,-48))
cice_output_file = netcdfpath+"{}_CICE_basic_outputs.nc".format(track_number)
var = 'dafsd_wave_ra'
ds = intFloeSize(ds, var)
# var = 'dafsd_newi_ra'
# ds = intFloeSize(ds, var)
ds
````

````python
def load_closest_ww3_file(data_dir_ww3, target_time_ww3):
    files_ww3 = sorted([f for f in os.listdir(data_dir_ww3) 
                    if f.startswith("ww3_om2") and f.endswith(".nc")])

    file_dates_ww3 = []
    for f in files_ww3:
        try:
            date_str_ww3 = f.split('_')[-1].split('.')[0]
            file_dates_ww3.append(pd.Timestamp(date_str_ww3))
        except Exception as e:
            print(f"Skipping file {f} due to error: {e}")
            continue

    file_dates_ww3 = [d for d in file_dates_ww3 if isinstance(d, pd.Timestamp)]
    closest_file_ww3 = files_ww3[min(range(len(file_dates_ww3)), key=lambda i: abs(file_dates_ww3[i] - target_time_ww3))]

    file_path_ww3 = os.path.join(data_dir_ww3, closest_file_ww3)
    ds_ww3 = xr.open_dataset(file_path_ww3)

    # print(f"Loading file: {closest_file_ww3}")
    return ds_ww3

time_indices = [crossing_idx]

data_dir_ww3 = netcdfpath
#netcdfpath+'CAWCR/' #'/g/data/ia40/cice-dirs/input/CICE_data/forcing/access-om2-025/CAWCR/2016/'
# Extract the target time from the dataset as a numpy datetime64 object and convert to pandas Timestamp
target_time_ww3 = pd.to_datetime(ds['time'].isel(time=time_indices[0]).values).floor('D')

ds_ww3 = load_closest_ww3_file(data_dir_ww3, target_time_ww3)
ds_ww3_snapshot = ds_ww3.sel(time=target_time_ww3)
ds_ww3_snapshot['time']

ds_ww3_snapshot
````
